```python
mixer.init()
sound = mixer.Sound('alarm.wav')

lbl=['banana','Cassava','coffeetree','corn','sugarcane']

model = load_model('models/cnnCat2Plants.h5')
path = os.getcwd()
```

# Code for detecting plants

```python
import cv2
import numpy as np

# Replace the following line with the path to your recorded video file
video_path = 'videos/Garden Video.mp4'
cap = cv2.VideoCapture(video_path)

font = cv2.FONT_HERSHEY_COMPLEX_SMALL
count = 0
score = 0
thicc = 2
rpred = [99]
lpred = [99]

# Define the resized frame dimensions
resized_width = 640
resized_height = 480

# Define the plant labels
#lbl = ['Cassava', 'Corn', 'Banana']

while(cap.isOpened()):
    ret, frame = cap.read()

    if not ret:
        break

    frame = cv2.resize(frame, (resized_width, resized_height))

    height, width = frame.shape[:2]

    # Plant detection code
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower_green = np.array([30, 50, 50])
    upper_green = np.array([70, 255, 255])
    mask = cv2.inRange(hsv, lower_green, upper_green)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 500:
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(frame, (x, y), (x+w, y+h), (100, 100, 100), 1)

            # Classification code for the detected plant region
            plant_region = frame[y:y+h, x:x+w]
            plant_region = cv2.cvtColor(plant_region, cv2.COLOR_BGR2GRAY)
            plant_region = cv2.resize(plant_region, (24, 24))
            plant_region = plant_region / 255
            plant_region = plant_region.reshape(24, 24, -1)
            plant_region = np.expand_dims(plant_region, axis=0)
            pred = model.predict(plant_region)
            print('pred value:', str(pred))
            print("np argmax: "+str(np.argmax(pred)))
            pred_label = lbl[np.argmax(pred)]

            cv2.putText(frame, pred_label, (x, y-10), font, 0.8, (0, 255, 0), 1, cv2.LINE_AA)

    # Display frame with results
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
